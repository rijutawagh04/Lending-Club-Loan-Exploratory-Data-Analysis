## Loading cleaned accepted loan data file into loan dataframe

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels
import pandasql

data_directory = os.path.join('.', 'CleanedData')

loan_data_path = os.path.join(data_directory, 'CleanedAcceptedLoanData.csv')
loan = pd.read_csv(loan_data_path, low_memory=False, encoding='latin-1')
```

### Wordcloud of purpose for which the loans were taken

```python
from os import path
import wordcloud
from wordcloud import WordCloud
plt.figure(figsize=(10,7))
text = loan['title'].to_json()
wc = WordCloud(ranks_only=True,prefer_horizontal = 0.6,background_color = 'white',
              max_words = 50).generate(text)
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
import seaborn as sns
import matplotlib

s = pd.value_counts(loan['emp_length']).to_frame().reset_index()
s.columns = ['type', 'count']

def emp_dur_graph(graph_title):

    sns.set_style("whitegrid")
    ax = sns.barplot(y = "count", x = 'type', data=s)
    ax.set(xlabel = 'Emploment Length(Years)', ylabel = 'No of issued loans', title = graph_title)
    ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    _ = ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
    
    
    
#emp_dur_graph('Distribution of employment length for issued loans')
plt.style.use('classic')
ax = emp_dur_graph('Distribution of employment length for issued loans')
plt.show()
```
